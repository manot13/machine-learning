```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/diabetes.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
df_sem_outcome = df.iloc[:, 0:8]
```

```python
df_sem_outcome
```

```python
preenche_0 = SimpleImputer(missing_values = 0, strategy = 'mean')
df_sem_outcome = preenche_0.fit_transform(df_sem_outcome)
```

```python
labels = list(df.iloc[0, 0:8].index[0:8])
```

```python
df2 = pd.DataFrame(df_sem_outcome, columns=[labels])
```

```python
df2
```

```python
df_outcome = df.iloc[:, 8]
```

```python
#df_outcome = pd.DataFrame(df_outcome, columns=['Outcome'])
```

```python
list(df_outcome)
```

```python
df2
```

```python
df_def = df2.assign(Outcome=list(df_outcome))
```

```python
df_def
```

```python
x_diabetes = df_def.iloc[:, 0:8].values
y_diabetes = df_def.iloc[:, 8].values
```

```python
x_diabetes.shape, y_diabetes.shape
```

```python
x_diabetes[0]
```

```python
from sklearn.preprocessing import StandardScaler
scaler_diabetes = StandardScaler()
x_diabetes = scaler_diabetes.fit_transform(x_diabetes)
```

```python
x_diabetes
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_diabetes_treinamento, x_diabetes_teste, y_diabetes_treinamento, y_diabetes_teste = train_test_split(x_diabetes, y_diabetes, test_size = 0.30, random_state = 42)
```

```python
x_diabetes_treinamento.shape, y_diabetes_treinamento.shape
```

```python
x_diabetes_teste.shape, y_diabetes_teste.shape
```

Com algoritmo Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
modelo = GaussianNB()
```

```python
modelo.fit(x_diabetes_treinamento, y_diabetes_treinamento)
```

```python
previsoes = modelo.predict(x_diabetes_teste)
previsoes
```

```python
y_diabetes_teste
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from yellowbrick.classifier import ConfusionMatrix
```

```python
accuracy_score(y_diabetes_teste, previsoes)
```

```python
cm = ConfusionMatrix(modelo)
cm.fit(x_diabetes_treinamento, y_diabetes_treinamento)
cm.score(x_diabetes_teste, y_diabetes_teste)
```

```python
print(classification_report(y_diabetes_teste, previsoes))
```

Com algoritmo DecisionTree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
arvore_diabetes = DecisionTreeClassifier(criterion='entropy', random_state=42)
arvore_diabetes.fit(x_diabetes_treinamento, y_diabetes_treinamento)
```

```python
previsoes2 = arvore_diabetes.predict(x_diabetes_teste)
previsoes2
```

```python
accuracy_score(y_diabetes_teste, previsoes2)
```

```python
cm2 = ConfusionMatrix(arvore_diabetes)
cm2.fit(x_diabetes_treinamento, y_diabetes_treinamento)
cm2.score(x_diabetes_teste, y_diabetes_teste)
```

Com algoritmo KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn_diabetes = KNeighborsClassifier(n_neighbors=10)
knn_diabetes.fit(x_diabetes_treinamento, y_diabetes_treinamento)
```

```python
previsoes3 = knn_diabetes.predict(x_diabetes_teste)
previsoes3
```

```python
accuracy_score(y_diabetes_teste, previsoes3)
```

```python
cm3 = ConfusionMatrix(knn_diabetes)
cm3.fit(x_diabetes_treinamento, y_diabetes_treinamento)
cm3.score(x_diabetes_teste, y_diabetes_teste)
```
